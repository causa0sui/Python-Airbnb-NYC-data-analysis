**Airbnb** is an online marketplace for arranging or offering lodging, primarily homestays, or tourism experiences. The company does not own any of the real estate listings, nor does it host events; it acts as a broker, receiving commissions from each booking. The company is based in San Francisco, California, United States.

Airbnb provides a platform for hosts to accommodate guests with short-term lodging and tourism-related activities. Guests can search for lodging using filters such as lodging type, dates, location, and price. Guests have the ability to search for specific types of homes, such as bed and breakfasts, unique homes, and vacation homes. 

Before booking, users must provide personal and payment information. Some hosts also require a scan of government-issued identification before accepting a reservation. Guests have the ability to chat with hosts through a secure messaging system. Hosts provide prices and other details for their rental or event listings, such as the allowed number of guests, home type, rules, and amenities. 
Pricing is determined by the host, with recommendations from Airbnb. Hosts and guests have the ability to leave reviews about the experience.


The purpose of this project is to explore and analyze Aibnb New York City data for the first half of 2019. 
We would like to inspect based on available data which part of Big Apple are the most popular among tourists who use this site to book an accommodation, which kinds of apartment are the most rented and whether average person coming to New York can afford a rent of the place based on different criteria with help of *classification* algorithms.

# Handling missing data and preparation for modelling

Now it is time to impute missing data to some of the variables which have them.

We impute the missing data for **reviews_per_month** with random numbers within the interval of plus-minus standard deviation span of mean.

Also we drop *host_id* because this variable is useless in our analysis.

```python
data = data.drop(["host_id"], axis = 1)
#host_id is not informative here
```

```python
#generate random numbers of size of null values withing range of mean +-standard deviation
#review per month
mean = data.reviews_per_month.mean()
std = data.reviews_per_month.std()
random_imputation = np.random.randint(mean - std, mean + std, size = data.reviews_per_month.isnull().sum())
#copy
copy = data.reviews_per_month.copy()
copy[np.isnan(copy)] = random_imputation
data.reviews_per_month = copy
data.reviews_per_month = data.reviews_per_month.astype(int)
data.reviews_per_month
data.head()
```

Now we handle missing data of the subsequent variables.

```python
#handling host_name and name missing values - it is enough to substitute them by the most common value
data["name"] = data.fillna(data.mode().iloc[0])
```

```python
data["host_name"].fillna("John", inplace = True)

data["afford"].isnull().sum()
```

```python
data.neighbourhood.describe()
#this variable is not informative as categorical, because it has 221 unique values, so we drop it
data = data.drop("neighbourhood", axis = 1)
```

# Data set split and subsequent Machine Learning algorithms 

Now we split our dataset into training and test part with test size of 30%, random state = 42.

```python
df = data.drop(['name', 'host_name', 'price_group',
'last_review', 'avail_during_year', 'min_price_stay', 
                'price'], axis=1)
df = df.dropna()
df_x = df.drop(["afford"], axis = 1)
df_y = df["afford"]
```

### Stochastic Gradient Descent

Next classifier to be tested is **Stochastic Gradient Descent** (SDG).

Gradient descent is an iterative algorithm, that starts from a random point on a function and travels down its slope in steps until it reaches the lowest point of that function.

There are a few downsides of the gradient descent algorithm. We need to take a closer look at the amount of computation we make for each iteration of the algorithm.


Say we have 10,000 data points and 10 features. The sum of squared residuals consists of as many terms as there are data points, so 10000 terms in our case. We need to compute the derivative of this function with respect to each of the features, so in effect we will be doing 10000 * 10 = 100 000 computations per iteration. It is common to take 1000 iterations, in effect we have 100 000 * 1000 = 100 000 000 computations to complete the algorithm. That is pretty much an overhead and hence gradient descent is slow on huge data.



It is while selecting data points at each step to calculate the derivatives. SGD randomly picks one data point from the whole data set at each iteration to reduce the computations enormously.

SGD has only **70%** accuracy on our dataset. But this value is varying depending on different ways of performing of the algorithm.

```python
#Stochastic Gradient Descent (SDG)
sgd = linear_model.SGDClassifier(max_iter=5, tol=None)
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)

sgd.score(X_train, Y_train)

acc_sgd = round(sgd.score(X_test, Y_test), 2)
acc_sgd
```
